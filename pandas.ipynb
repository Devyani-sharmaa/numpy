```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
a=np.array([1,4,6,4,3,7,8])
```

```python
b=np.array([2,4,6,7,5,4,3])
```

```python
plt.plot(a,b)
```

# Marker

```python
plt.plot(a,marker="*")
```

```python
plt.plot(b,marker='>',ms=10)           #ms=markersize
```

```python
plt.plot(a, b,marker='<',mec='green',mfc='yellow',ms=10)

for i in range(len(a)):
    plt.text(a[i],b[i],f"({a[i]},{b[i]})")
```

# Labels

```python
plt.plot(a, b,marker='<',mec='green',mfc='yellow',ms=10)

for i in range(len(a)):
    plt.text(a[i],b[i],f"({a[i]},{b[i]})")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("info",color="red",size=30)
plt.grid()  # to make lines in background to understand graph esay
```

```python
plt.plot(a, b,marker='<',mec='green',mfc='yellow',ms=10)

for i in range(len(a)):
    plt.text(a[i],b[i],f"({a[i]},{b[i]})")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("info",color="red",size=30)
plt.grid(axis="y")  # to make lines in  y axis only 
```

# #Lines--- > we can provide color to line and linestyle have 3 method --dotted,dash,dashdot

```python
plt.plot(a, b,marker='<',mec='green',mfc='yellow',ms=10,color="red",linewidth=3,linestyle="dotted")

for i in range(len(a)):
    plt.text(a[i],b[i],f"({a[i]},{b[i]})")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("info",color="red",size=30)
plt.grid()  # to make lines in background to understand graph esay
```

```python
x=np.array([12,34,5,6,77,89,98,75,78])
y=np.array([22,54,65,76,42,54,78,78,54])
z=np.array([32,54,76,43,65,76,87,65,89])
plt.figure(facecolor="pink")
plt.plot(x,z,label="age",marker="*")
plt.plot(y,z,label='salary',marker=">")
plt.legend()
```

# SCATTER PLOT

```python
x=np.array([25,45,55,75,35])
y=np.array([54,72,64,96,24])

a=np.array([34,56,32,32,67])
b=np.array([36,54,43,78,54])
plt.scatter(x,y,color='red',label='profit')
plt.scatter(a,b,color="pink",label='loss')
plt.legend(loc='center')

for i in range(len(x)):
    plt.text(x[i],y[i],f"({x[i]},{y[i]})",color='blue')
for i in range(len(a)):
    plt.text(a[i],b[i],f"({a[i]},{b[i]})",color='brown')
    
```

# scatter using variables

```python
x=np.array([10,20,30,40,50])
y=np.array([15,20,25,30,45])

color=["red","blue","green","pink",'grey']
size=[200,300,20,400,90]
plt.scatter(x,y,color=color,sizes=size,label="loss")
plt.grid()
plt.legend(loc='upper left')
plt.xlabel("bkj")
plt.ylabel("profit")
```
